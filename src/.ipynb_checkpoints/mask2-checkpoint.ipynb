```python
# Imports
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Define o caminho e inicializa a imagem
face = '../testimages/img_1.png'
image = Image.open(face)
image = np.asarray(image)
```

```python
plt.imshow(image)
```

```python
hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
grayscaled = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayscaled, cmap="gray")
```

```python
blur = cv2.ximgproc.l0Smooth(grayscaled)
plt.imshow(blur, cmap="gray")

A = np.amax(blur)
ret, thresh = cv2.threshold(blur, A-20, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap="gray")
```

```python

print(A)

edges=cv2.Canny(blur, 5, 15, 50)
plt.imshow(edges, cmap="gray")
```
